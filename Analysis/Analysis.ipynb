```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample_data/Time_CKD.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Algorithms'], df['Time'])
plt.xlabel('Algorithm')
plt.ylabel('Time(sec)')
plt.show()
```

```python
Algorithms = df['Algorithms'].values.tolist()
for i in range(len(Algorithms)):
  x = Algorithms[i].split(' ')
  if len(x) ==2:
    try:
      x.insert(x.index('LSTM'), '\n')
    except:
      x.insert(x.index('GRU'), '\n')
  else:
    # x.insert(x[0], '\n')
    x[0] += '\n'
  Algorithms[i] = ' '.join(x)
plt.bar(Algorithms, df['Time'])
plt.tick_params(axis='x', which='major', labelsize=8)
plt.tight_layout()
```
